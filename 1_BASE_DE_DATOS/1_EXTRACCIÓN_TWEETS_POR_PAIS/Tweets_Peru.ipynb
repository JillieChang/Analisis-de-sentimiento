## EXTRACCIÓN DE TWEETS DE PERÚ

### 1. Cargar librerías paquetes 

```python
import pandas as pd
import numpy as np
import GetOldTweets3 as got
import time
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Extraer datos

Se extraen los datos de enero a marzo de 2020

```python
text_query = 'covid OR coronavirus OR pandemia OR cuarentena'
since_date = "2020-1-01"
until_date = "2020-4-01"
geocode = "-12.101026, -77.068738" #Lima, Peru"
km = "46.72934km"
tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setNear(geocode).setWithin(km).setSince(since_date).setUntil(until_date).setEmoji("unicode")
tweets = got.manager.TweetManager.getTweets(tweetCriteria)
tweets_list = [[tweet.id, tweet.author_id, tweet.username, tweet.text, tweet.retweets, tweet.favorites, tweet.replies, tweet.date, tweet.hashtags] for tweet in tweets]
 # Creation of dataframe from tweets list # Add or remove columns as you remove tweet information
tweets_peru1 = pd.DataFrame(tweets_list, columns = ['Tweet_Id', 'Tweet_User_Id', 'Tweet_User', 'Text',
                                                  'Retweets', 'Favorites', 'Replies', 'Datetime', 'hashtags'])
```

```python
tweets_peru1.to_csv('covid_peru_emoji1.csv', header=True, index=False, encoding='utf-8-sig')
```

Se extraen los datos de abril a agosto de 2020

```python
text_query = 'covid OR coronavirus OR pandemia OR cuarentena'
since_date = "2020-4-01"
until_date = "2020-9-01"
geocode = "-12.101026, -77.068738" #Lima, Peru"
km = "46.72934km"
tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setNear(geocode).setWithin(km).setSince(since_date).setUntil(until_date).setEmoji("unicode")
tweets = got.manager.TweetManager.getTweets(tweetCriteria)
tweets_list = [[tweet.id, tweet.author_id, tweet.username, tweet.text, tweet.retweets, tweet.favorites, tweet.replies, tweet.date, tweet.hashtags] for tweet in tweets]
 # Creation of dataframe from tweets list # Add or remove columns as you remove tweet information
tweets_peru2 = pd.DataFrame(tweets_list, columns = ['Tweet_Id', 'Tweet_User_Id', 'Tweet_User', 'Text',
                                                  'Retweets', 'Favorites', 'Replies', 'Datetime', 'hashtags'])
```

```python
tweets_peru2.to_csv('covid_peru_emoji2.csv', header=True, index=False, encoding='utf-8-sig')
```

### 3. Se guarda la base

Se juntan las dos bases

```python
PERU_CONCAT = pd.concat([tweets_peru1, tweets_peru2])
PERU_CONCAT.shape
```

Se eliminan duplicados

```python
PERU_CONCAT_DISTINCT= PERU_CONCAT.drop_duplicates()
```

```python
PERU_CONCAT_DISTINCT.reset_index(drop=True, inplace=True)
```

```python
PERU_CONCAT_DISTINCT.to_csv('PERU_CONCAT_DISTINCT.csv', header=True, index=False, encoding='utf-8-sig')
```
